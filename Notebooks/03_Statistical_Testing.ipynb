```python
import pandas as pd
from scipy import stats

df = pd.read_csv('03_Data_cleaned_after_pandas.csv')
df.head(2)
```

```python
contingency = pd.crosstab(df['OverTime'], df['Attrition'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"P-Value: {p:.6f}")
print("Result: SIGNIFICANT" if p < 0.05 else "Result: NOT SIGNIFICANT")
```

```python
left = df[df['Attrition']=='Yes']['MonthlyIncome']
stayed = df[df['Attrition']=='No']['MonthlyIncome']
t_stat, p_val = stats.ttest_ind(left, stayed)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-Value: {p_val:.6f}")
print(f"Mean income (Left): {left.mean():.0f}")
print(f"Mean income (Stayed): {stayed.mean():.0f}")
```
